```python
import pandas as pd
import pymysql
from sqlalchemy import create_engine
```

```python
#load the datastet
df = pd.read_csv("C:/Users/91801/Desktop/Agriculture/Agriculture analysis/Agriculture Data.csv")
```

```python
#rename the columns
df.columns = df.columns.str.lower()
```

```python
#replace the spacse with underscore
df.columns = df.columns.str.replace(" ","_")
```

```python
#use regex function to check unvanted replace that
df.columns = df.columns.str.replace(r"_\(.*\)", "", regex=True)
```

```python
#check if any null values
df.isnull().sum()/len(df)
```

```python
#check if any duplictes value
df.duplicated().sum()
```

```python
#filter the columns with specific business cases
agri_data = df[[
    "dist_code", "state_code", "state_name", "dist_name", "year",
    "rice_area", "rice_production", "rice_yield",
    "wheat_area", "wheat_production", "wheat_yield",
    "kharif_sorghum_area", "kharif_sorghum_production",
    "rabi_sorghum_area", "rabi_sorghum_production",
    "sorghum_production", "maize_area", "maize_production",
    "maize_yield", "groundnut_production", "soyabean_area",
    "soyabean_production","soyabean_yield", "oilseeds_production", 
    "oilseeds_area","sunflower_production", "sugarcane_production", 
    "cotton_production", "cotton_yield", "pearl_millet_production","finger_millet_production"]]

agri_data
```

```python
#connect the Mysql and push the DataFrame in Mysql
def create_database_and_table(df, user, password, host, port, database, table_name):
    try:
        myconnect = pymysql.connect(host=host, user=user, passwd=password)
        cursor = myconnect.cursor()
        
        cursor.execute(f"CREATE DATABASE IF NOT EXISTS {database}")
        print(f"✅ Database '{database}' is ready.")

        engine = create_engine(f'mysql+pymysql://{user}:{password}@{host}:{port}/{database}')
        
        df.to_sql(name=table_name, con=engine, if_exists='replace', index=False)
        print(f"✅ Table '{table_name}' created and data inserted successfully.")

    except Exception as e:
        print(f"⚠️ Error: {e}")
        
create_database_and_table(agri_data, "root", "0000", "127.0.0.1", "3306", "Agriculture", "agri_data")
```
